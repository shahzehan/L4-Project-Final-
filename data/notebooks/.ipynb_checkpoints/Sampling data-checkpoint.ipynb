```python
#import config

import gzip
import shutil
import os
import wget
import csv
import linecache
from shutil import copyfile
import ipywidgets as widgets
import numpy as np
import pandas as pd
import dask.dataframe as dd
```

```python
#Downloads the dataset (compressed in a GZ format)
wget.download("https://zenodo.org/record/5652342/files/full_dataset_clean.tsv.gz?download=1", out = '..\\raw\\full_dataset_clean.tsv.gz')

#Unzips the dataset and gets the TSV dataset
with gzip.open('..\\raw\\full_dataset_clean.tsv.gz', 'rb') as f_in:
    with open('..\\raw\\full_dataset_clean.tsv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

Deletes the compressed GZ file
os.unlink("..\\raw\\full_dataset_clean.tsv.gz")

df = pd.read_csv('..\\raw\\full_dataset_clean.tsv', sep = '\t')
df.head()
```

```python
# Coverting date to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
jan = df[(df['lang'] == "en") & (df['date'] >= '2020-01-01') & (df['date'] < '2020-02-01')]
jan.head()
```

```python
jan.info()
```

```python
jan = jan.sample(n = 150000)
jan.info()
```

```python
jan1 = jan.iloc[:len(jan) // (3/2)]
jan1 = jan1['tweet_id']

jan2 = jan.iloc[len(jan) // (3/2):]
jan2 = jan2['tweet_id']

jan1.to_csv('..\\processed\\Jan_2020A.csv', header = False, index = False,)
jan2.to_csv('..\\processed\\Jan_2020B.csv', header = False, index = False)
```

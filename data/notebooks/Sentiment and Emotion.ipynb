```python
def plot_sen(df, period, image_name):
    dfa = df[df['topic'] == "Masks"]
    dfap = round((dfa['sentiment'].value_counts()['POS'] / dfa.shape[0]) * 100, 2)
    dfan = round((dfa['sentiment'].value_counts()['NEG'] / dfa.shape[0]) * 100, 2)
    dfau = round((dfa['sentiment'].value_counts()['NEU'] / dfa.shape[0]) * 100, 2)

    dfb = df[df['topic'] == "Symptoms"]
    dfbp = round((dfb['sentiment'].value_counts()['POS'] / dfb.shape[0]) * 100, 2)
    dfbn = round((dfb['sentiment'].value_counts()['NEG'] / dfb.shape[0]) * 100, 2)
    dfbu = round((dfb['sentiment'].value_counts()['NEU'] / dfb.shape[0]) * 100, 2)

    dfc = df[df['topic'] == "Treatment"]
    dfcp = round((dfc['sentiment'].value_counts()['POS'] / dfc.shape[0]) * 100, 2)
    dfcn = round((dfc['sentiment'].value_counts()['NEG'] / dfc.shape[0]) * 100, 2)
    dfcu = round((dfc['sentiment'].value_counts()['NEU'] / dfc.shape[0]) * 100, 2)

    dfd = df[df['topic'] == "Science"]
    dfdp = round((dfd['sentiment'].value_counts()['POS'] / dfd.shape[0]) * 100, 2)
    dfdn = round((dfd['sentiment'].value_counts()['NEG'] / dfd.shape[0]) * 100, 2)
    dfdu = round((dfd['sentiment'].value_counts()['NEU'] / dfd.shape[0]) * 100, 2)

    dfe = df[df['topic'] == "Health"]
    dfep = round((dfe['sentiment'].value_counts()['POS'] / dfe.shape[0]) * 100, 2)
    dfen = round((dfe['sentiment'].value_counts()['NEG'] / dfe.shape[0]) * 100, 2)
    dfeu = round((dfe['sentiment'].value_counts()['NEU'] / dfe.shape[0]) * 100, 2)

    dff = df[df['topic'] == "Economy"]
    dffp = round((dff['sentiment'].value_counts()['POS'] / dff.shape[0]) * 100, 2)
    dffn = round((dff['sentiment'].value_counts()['NEG'] / dff.shape[0]) * 100, 2)
    dffu = round((dff['sentiment'].value_counts()['NEU'] / dff.shape[0]) * 100, 2)
    
    pos = [dfap, dfbp, dfcp, dfdp, dfep, dffp]
    neg = [dfan, dfbn, dfcn, dfdn, dfen, dffn]
    neu = [dfau, dfbu, dfcu, dfdu, dfeu, dffu]
    
    n=6
    r = np.arange(n)
    w = 0.24
    
    plt.bar(r, pos, color='#4594d0', width=w, edgecolor='black', label='Positive')
    plt.bar(r+w, neg, color='#ecb77e', width=w, edgecolor='black', label='Negative')
    plt.bar(r+w+w, neu, color='#c3ffa8', width=w, edgecolor='black', label='Neutral')
    
    plt.xlabel("Time Period")
    plt.ylabel("Sentiment Ration (%)")
    plt.title("{} Ratios of Sentiment per Topic".format(period))
    plt.xticks(r + w,['Masks','Symptoms','Treatment','Science','Health','Economy'])
    plt.legend()
    
    plt.savefig("..\\processed\\images\\{}.png".format(image_name), dpi=100)
    
def df2list(df):
    va = df['emotion'].value_counts()['joy']
    vb = df['emotion'].value_counts()['sadness']
    vc = df['emotion'].value_counts()['anger']
    vd = df['emotion'].value_counts()['surprise']
    ve = df['emotion'].value_counts()['disgust']
    vf = df['emotion'].value_counts()['fear']
    out = [va, vb, vc, vd, ve, vf]
    return out
    
def plot_emo(df, period, image_name):
    df = df[df['label'] == 1]
    dfa = df[df['topic'] == "Masks"]
    dfb = df[df['topic'] == "Symptoms"]
    dfc = df[df['topic'] == "Treatment"]
    dfd = df[df['topic'] == "Science"]
    dfe = df[df['topic'] == "Health"]
    dff = df[df['topic'] == "Economy"]
    
    emo1 = df2list(dfa)
    emo2 = df2list(dfb)
    emo3 = df2list(dfc)
    emo4 = df2list(dfd)
    emo5 = df2list(dfe)
    emo6 = df2list(dff)
    
    explode = [0.01, 0.01, 0.01, 0.01, 0.01, 0.01]
    labels = ['Joy', 'Sadness', 'Anger', 'Surprise', 'Disgust', 'Fear']
    
    fig, axes = plt.subplots(2, 3, figsize=(20, 8))
    
    axes[0, 0].pie(emo1, labels=labels, explode=explode, autopct='%.0f%%')
    axes[0, 0].set_title("Masks")
    
    axes[0, 1].pie(emo2, labels=labels, explode=explode, autopct='%.0f%%')
    axes[0, 1].set_title("Symptoms")
                     
    axes[0, 2].pie(emo3, labels=labels, explode=explode, autopct='%.0f%%')
    axes[0, 2].set_title("Treatment")
    
    axes[1, 0].pie(emo4, labels=labels, explode=explode, autopct='%.0f%%')
    axes[1, 0].set_title("Science")
    
    axes[1, 1].pie(emo5, labels=labels, explode=explode, autopct='%.0f%%')
    axes[1, 1].set_title("Health")
    
    axes[1, 2].pie(emo6, labels=labels, explode=explode, autopct='%.0f%%')
    axes[1, 2].set_title("Economy")
    
    plt.suptitle("{} Distribution of Emotions per Topic".format(period))
    plt.savefig("..\\processed\\images\\{}.png".format(image_name), dpi=100)
    
```

```python
df = df = pd.read_csv("..\\processed\\tweets\\Jan2020.csv", names=('text', 'label', 'topic', 'sentiment', 'emotion'))
plot_sen(df, "January 2020", "jan2020sen")
plot_emo(df, "January 2020", "jan2020emo")
```

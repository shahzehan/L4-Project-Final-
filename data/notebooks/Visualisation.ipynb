```python
# Visualisation Functions
topics = ["Masks", "Symptoms", "Treatment", "Science", "Health", "Economy"] 

# Misinformation ratio
def misinfo_ratio(m_df):
    dfa = m_df[m_df['topic'] == "Masks"]
    dfat = dfa['label'].value_counts()[0]
    dfaf = dfa['label'].value_counts()[1]
    dfar = dfaf / (dfat + dfaf)

    dfb = m_df[m_df['topic'] == "Symptoms"]
    dfbt = dfb['label'].value_counts()[0]
    dfbf = dfb['label'].value_counts()[1]
    dfbr = dfbf / (dfbt + dfbf)

    dfc = m_df[m_df['topic'] == "Treatment"]
    dfct = dfc['label'].value_counts()[0]
    dfcf = dfc['label'].value_counts()[1]
    dfcr = dfcf / (dfct + dfcf)

    dfd = m_df[m_df['topic'] == "Science"]
    dfdt = dfd['label'].value_counts()[0]
    dfdf = dfd['label'].value_counts()[1]
    dfdr = dfdf / (dfdt + dfdf)

    dfe = m_df[m_df['topic'] == "Health"]
    dfet = dfe['label'].value_counts()[0]
    dfef = dfe['label'].value_counts()[1]
    dfer = dfef / (dfet + dfef)

    dff = m_df[m_df['topic'] == "Economy"]
    dfft = dff['label'].value_counts()[0]
    dfff = dff['label'].value_counts()[1]
    dffr = dfff / (dfft + dfff)
    
    m_dft = m_df['label'].value_counts()[0]
    m_dff = m_df['label'].value_counts()[1]
    m_dfr = m_dff / (m_dft + m_dff)
    
    return [dfar, dfbr, dfcr, dfdr, dfer, dffr, m_dfr]

# Diff between misinformation between time periods
def ratio_diff(list_a, list_b):
    diff_list = []
    zip_obj = zip(list_a, list_b)
    for v1, v2 in zip_obj:
        diff_list.append(round((v1 - v2) / v2 * 100, 2))
        
    return diff_list

def tm_graphs(df, df1, df2, mi1, mi2, period, image_name):
    topics = ["Masks", "Symptoms", "Treatment", "Science", "Health", "Economy", "Overall"]
    df1_tweet_num = [df.iloc[mi1, 1], df.iloc[mi1, 3], df.iloc[mi1, 7], df.iloc[mi1, 8], df.iloc[mi1, 10], df.iloc[mi1, 11], sum(df.iloc[mi1, 1:16])]
    df2_tweet_num = [df.iloc[mi2, 1], df.iloc[mi2, 3], df.iloc[mi2, 7], df.iloc[mi2, 8], df.iloc[mi2, 10], df.iloc[mi2, 11], sum(df.iloc[mi2, 1:16])]
    
    tweet_num_diff = []
    
    for i in range(7):
        tweet_num_diff.append(round((df1_tweet_num[i] - df2_tweet_num[i]) / df2_tweet_num[i] * 100, 2))
    
    r1 = misinfo_ratio(df1)
    r2 = misinfo_ratio(df2)
    r_diff = ratio_diff(r1, r2)
    
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    
    
    tm_df = pd.DataFrame({"Topics":topics, "TweetNum":tweet_num_diff, "RatioDiff":r_diff})
    
    plot1 = sns.barplot(x="Topics", y="TweetNum", data=tm_df, color='#4594d0', ax=axes[0])
    for bar in plot1.patches:
        plot1.annotate(format(bar.get_height()), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='center', size=15, xytext=(0, 8), textcoords='offset points')
        
    axes[0].set_xlabel("Topics")
    axes[0].set_ylabel("Relative Change from Previous Time Period (%)")
    axes[0].set_title("{} Change in Tweets per Topic".format(period))
    axes[0].grid()
    
    plot2 = sns.barplot(x="Topics", y="RatioDiff", data=tm_df, color='#ecb77e', ax=axes[1])
    for bar in plot2.patches:
        plot2.annotate(format(bar.get_height()), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='center', size=15, xytext=(0, 8), textcoords='offset points')
        
    axes[1].set_xlabel("Topics")
    axes[1].set_ylabel("Relative Change from Previous Time Period (%)")
    axes[1].set_title("{} Change in Ratio of Misinformation per Topic".format(period))
    axes[1].grid()
    
    plt.savefig("..\\processed\\images\\{}.png".format(image_name), dpi=100)
    
```

```python
def plot_topic_emo(topic, df_list, clist):
    joy = []
    sad = []
    anger = []
    surp = []
    disg = []
    fear = []
    mis = []
    
    for i in range(len(df_list)):
        df_list[i] = df_list[i][(df_list[i]['topic'] == topic) & (df_list[i]['emotion'] != 'others')]
        joy.append(round(df_list[i]['emotion'].value_counts()['joy'] / df_list[i].shape[0], 3))
        sad.append(round(df_list[i]['emotion'].value_counts()['sadness'] / df_list[i].shape[0], 3))
        anger.append(round(df_list[i]['emotion'].value_counts()['anger'] / df_list[i].shape[0], 3))
        surp.append(round(df_list[i]['emotion'].value_counts()['surprise'] / df_list[i].shape[0], 3))
        disg.append(round(df_list[i]['emotion'].value_counts()['disgust'] / df_list[i].shape[0], 3))
        fear.append(round(df_list[i]['emotion'].value_counts()['fear'] / df_list[i].shape[0], 3))
        mis.append(round(df_list[i]['label'].value_counts()[1] / df_list[i].shape[0], 3))
        
    time_periods = ["Jan20", "Apr20", "Jul20", "Oct20", "Jan21", "Apr21", "Jul21", "Oct21"]
    
    plt.plot(time_periods, joy, label="Joy")
    plt.plot(time_periods, sad, label="Sadness")
    plt.plot(time_periods, anger, label="Anger")
    plt.plot(time_periods, surp, label="Surprise")
    plt.plot(time_periods, disg, label="Disgust")
    plt.plot(time_periods, fear, label="Fear")
    plt.plot(time_periods, mis, label="Misinformation")
    
    plt.xlabel("Time Periods")
    plt.ylabel("Ratio of Emotions")
    plt.title("Ratio of Emotions Over Time for {}".format(topic))
    plt.grid(True)
    plt.legend()
    plt.show
    
    plt.savefig("..\\processed\\images\\{}Emotion.png".format(topic), dpi=100)
    plt.cla()
    plt.clf()
    
    c1 = round(s.pearsonr(mis, joy)[0], 3)
    c2 = round(s.pearsonr(mis, sad)[0], 3)
    c3 = round(s.pearsonr(mis, anger)[0], 3)
    c4 = round(s.pearsonr(mis, surp)[0], 3)
    c5 = round(s.pearsonr(mis, disg)[0], 3)
    c6 = round(s.pearsonr(mis, fear)[0], 3)
    clist.append([c1, c2, c3, c4, c5, c6])
    
```

```python
def plot_topic_sen(topic, df_list, clist):
    pos = []
    neg = []
    neu = []
    mis = []
    
    for i in range(len(df_list)):
        df_list[i] = df_list[i][df_list[i]['topic'] == topic]
        pos.append(round(df_list[i]['sentiment'].value_counts()['POS'] / df_list[i].shape[0], 3))
        neg.append(round(df_list[i]['sentiment'].value_counts()['NEG'] / df_list[i].shape[0], 3))
        neu.append(round(df_list[i]['sentiment'].value_counts()['NEU'] / df_list[i].shape[0], 3))
        mis.append(round(df_list[i]['label'].value_counts()[1] / df_list[i].shape[0], 3))
        
    time_periods = ["Jan20", "Apr20", "Jul20", "Oct20", "Jan21", "Apr21", "Jul21", "Oct21"]
    
    plt.plot(time_periods, pos, label="Positive")
    plt.plot(time_periods, neg, label="Negative")
    plt.plot(time_periods, neu, label="Neutral")
    plt.plot(time_periods, mis, label="Misinformation")
    
    plt.xlabel("Time Periods")
    plt.ylabel("Ratio of Sentiments")
    plt.title("Ratio of Sentiments Over Time for {}".format(topic))
    plt.grid(True)
    plt.legend()
    plt.show
    
    plt.savefig("..\\processed\\images\\{}Sentiment.png".format(topic), dpi=100)
    plt.cla()
    plt.clf()
    
    c1 = round(s.pearsonr(mis, pos)[0], 3)
    c2 = round(s.pearsonr(mis, neg)[0], 3)
    c3 = round(s.pearsonr(mis, neu)[0], 3)
    clist.append([c1, c2, c3])
```

```python
sen_avg = []
emo_avg = []
topic_avg = []
tsen_avg = []

for i in range(3):
    ssum = 0
    for j in range(8):
        ssum += clist_sen[j][i]
    sen_avg.append(round(ssum / 8, 3))
    
for i in range(6):
    esum = 0
    for j in range(8):
        esum += clist_emo[j][i]
    emo_avg.append(round(esum / 8, 3))
    
for i in range(8):
    tsum = 0
    tsum2 = 0
    for j in range(6):
        tsum += clist_emo[i][j]
    for k in range(3):
        tsum2 += clist_sen[i][k]
    topic_avg.append(round(tsum / 6, 3))
    tsen_avg.append(round(tsum2 / 3, 3))
```

```python
print(sen_avg)
print(emo_avg)
print(topic_avg)
print(tsen_avg)
```
